# Seasonal and Weather Impact

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(df['Date'])
# Extract the month or season from the date of each crime
# You can choose to extract either the month or the season based on your analysis requirements
df['Month'] = df['Date'].dt.month
```

```python
# 1->Winter, 2-> Summer, 3-> Spring, 4-> Fall
def month_to_season(month):
    if month in [12, 1, 2]:
        return 1
    elif month in [3, 4, 5]:
        return 2
    elif month in [6, 7, 8]:
        return 3
    else:
        return 4

df['Season'] = df['Date'].dt.month.apply(month_to_season)
print(df[['Date', 'Season']].head())
```

```python
# Group the data by season and crime type to analyze crime frequency
crime_by_season = df.groupby(['Season', 'Primary Type']).size().unstack(fill_value=0)

# Print the first few rows to understand the structure of the grouped data
print(crime_by_season.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize crime frequency by season using a heatmap or line plot
plt.figure(figsize=(12, 8))
sns.heatmap(crime_by_season, cmap="YlGnBu", linewidths=.5)
plt.title('Crime Frequency by Season and Crime Type')
plt.xlabel('Crime Type')
plt.ylabel('Season')
plt.show()
```

```python
# detecting the skewed columns using plot
def plot(data,column):
  #distplot
  plt.figure(figsize=(15,4))
  plt.subplot(1,3,1)
  sns.distplot(data[column])
  plt.title("distplot for "+column)

  #histogram plot

  plt.subplot(1,3,2)
  sns.histplot(data, x= column, kde= True, bins=30,color="salmon")
  plt.title("histogram plot for "+column)

  #boxplot

  plt.subplot(1,3,3)
  sns.boxplot(data, x=column)
  plt.title("Box plot for "+column)
```
